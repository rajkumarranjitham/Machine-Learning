```python
import pandas as pd#importing pandas library and this is for to handling dataset and format is in table.
dataset=pd.read_csv("rice.csv")#reading the csv file and storing into dataset and will be in table format
dataset
```

```python
independant=dataset[["full_meals_order"]]#assinging the input x to the variable independant
independant
```

```python
dependant=dataset[["rice_in_kg"]]#assigning the output data y to the variable dependant
dependant
```

```python
from sklearn.model_selection import train_test_split#sklearn is a ML library and access model_selection and importing train_test_split class.
x_train,x_test,y_train,y_test=train_test_split(independant,dependant,test_size=0.3,random_state=0)#passing x,y,sie of train and test to the train_test_split and 
#assigning to x-train,y-train,x-test and y-test
x_train,x_test,y_train,y_test
```

```python
#making model learning by using sklearn library by using LinearRegression class and creating regressor variable
#by calling fit method make model to learn with train set
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_
weight
```

```python
bias=regressor.intercept_
bias
```

```python
#testing the model with y test data and saving the prediction data 
y_predict=regressor.predict(x_test)
y_predict
```

```python
#Evoluating metrics by using sklearn linrary and importing r2 class.
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_predict)
r_score
```

```python
#here validating the best model before deployment to live.
#by using pickel library dumping and writing the regressor data into pickle file with extenstion *.sav.
import pickle
filename="inventory_model.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#here loading the pickle data into the deployment model
deploy=pickle.load(open(filename,'rb'))
```

```python
check=deploy.predict([[100]])#cross-verifying the test data set
check
```
