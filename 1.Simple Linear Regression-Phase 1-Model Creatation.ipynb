```python
#pandas is a library and it contains the functions related to the database
#It handles the data in table format
import pandas as pd
```

```python
dataset=pd.read_csv("Salary_Data.csv")#dataset is created in Pandas by using method 'read_csv()'
```

```python
dataset #checking the dataset created in Pandas
```

```python
independent=dataset[["YearsExperience"]]#Assigning input 'x' to an independent variable
independent
```

```python
dependent=dataset[["Salary"]] #Assiging output 'y' to a dependent variable
dependent
```

```python
from sklearn.model_selection import train_test_split #sklearn is a ML library and importing train_test_split class
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0) 
#Passing arguments x,y,test_size (test percentage as 30 and 70% for learning) and random_state attibutes to the train_test_split method.
#Assigning to X_train as 70% of the independent , x_test as 70% of the dependent.
#Assigning to Y_train as 30% of the dependent , y_test as 30% of the dependent.

```

```python
from sklearn.linear_model import LinearRegression#from sklearn calling linear model library and importing linear regression class.
regressor=LinearRegression()#assiging Linear Regression() class behaviors to the regressor variable
regressor.fit(X_train,y_train)#calling regressor default method fit and passing splited input train and ouput train.
```

```python
weight=regressor.coef_ #calling coef_ (co-effeicient) method from regressor class to get the w or m or slope value and assinging to weight variable.
weight
```

```python
bais=regressor.intercept_ #calling intercept_ method from regressor to get the bias/initial value and assigning to bias variable.
bais
```

```python
y_pred=regressor.predict(X_test)#passing 30% of input test to the predict method of regressor and assiging to y_pred variable or output prediction.
```

```python
from sklearn.metrics import r2_score#calling evalution metrics from the sklearn folder and importing assesement metrics of r square_score.
r_score=r2_score(y_test,y_pred)#passing actual output and predicted ouput to the r square_score method and assigning to r_score variable.
```

```python
r_score#viewing the score
```

```python
import pickle#picle is a library here using for to save the best model.
filename="finalized_model_linear.sav"#creating a file named filename and assinging with besst model as *.sav
```

```python
pickle.dump(regressor,open(filename,'wb'))#writing binary of the best model from regression into pickel filename
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))#deploying best model by loading into pickel by reading binary of the best model.
result=loaded_model.predict([[13]])#passing input to the predict method of loaded model.
```
